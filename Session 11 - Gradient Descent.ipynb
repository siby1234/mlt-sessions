## Sourcing the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
oly=pd.read_csv('olympics.csv')
```

```python
oly.head()
```

```python
oly
```

## preprocessing

```python
X=oly['Year']
X
```

```python
y=oly['Time']
```

```python
y
```

```python
# Drawing a scatter diagram

plt.figure(figsize=(10,8))
plt.scatter(X,y);
```

## Designing a simple LR

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()

lr.fit(X.to_numpy().reshape(-1,1),y)
```

```python
# Plotting the LR line

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)));
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

p2=PolynomialFeatures(degree=2)

Xp2=p2.fit_transform(X.to_numpy().reshape(-1,1))
Xp2
```

```python
X
```

```python
lr2=LinearRegression()
pr2=lr2.fit(Xp2,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)))
plt.plot(X,pr2.predict(Xp2));
```

```python
## Polynomial of degree 3

p3= PolynomialFeatures(degree=3)
Xp3=p3.fit_transform(X.to_numpy().reshape(-1,1))
Xp3
lr3=LinearRegression()
pr3=lr3.fit(Xp3,y)
```

```python
# Plotting 

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='St line')
plt.plot(X,pr2.predict(Xp2),label='Poly of degree 2')
plt.plot(X, pr3.predict(Xp3), label='Poly of degree 3')
plt.legend();
```

```python
## Poly of degree 5

p5=PolynomialFeatures(degree=5)
Xp5=p5.fit_transform(X.to_numpy().reshape(-1,1))
lr5=LinearRegression()
pr5=lr5.fit(Xp5,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='St line')
plt.plot(X,pr2.predict(Xp2),label='Poly of degree 2')
plt.plot(X, pr3.predict(Xp3), label='Poly of degree 3')
plt.plot(X,pr5.predict(Xp5),label= 'Poly of degree 5')
plt.legend();
```

```python
## Poly of degree 10

p10=PolynomialFeatures(degree=10)
Xp10=p10.fit_transform(X.to_numpy().reshape(-1,1))
lr10=LinearRegression()
pr10=lr10.fit(Xp10,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='St line')
plt.plot(X,pr2.predict(Xp2),label='Poly of degree 2')
plt.plot(X, pr3.predict(Xp3), label='Poly of degree 3')
plt.plot(X,pr5.predict(Xp5),label= 'Poly of degree 5')
plt.plot(X,pr10.predict(Xp10),label= 'Poly of degree 10')
plt.legend();
```

```python
## Poly of degree 50

p50=PolynomialFeatures(degree=50)
Xp50=p50.fit_transform(X.to_numpy().reshape(-1,1))
lr50=LinearRegression()
pr50=lr50.fit(Xp50,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='St line')
plt.plot(X,pr2.predict(Xp2),label='Poly of degree 2')
plt.plot(X, pr3.predict(Xp3), label='Poly of degree 3')
plt.plot(X,pr5.predict(Xp5),label= 'Poly of degree 5')
plt.plot(X,pr10.predict(Xp10),label= 'Poly of degree 10')
plt.plot(X,pr50.predict(Xp50),label= 'Poly of degree 50')
plt.legend();
```

```python
## Poly of degree 75

p75=PolynomialFeatures(degree=75)
Xp75=p75.fit_transform(X.to_numpy().reshape(-1,1))
lr75=LinearRegression()
pr75=lr75.fit(Xp75,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='St line')
plt.plot(X,pr2.predict(Xp2),label='Poly of degree 2')
plt.plot(X, pr3.predict(Xp3), label='Poly of degree 3')
plt.plot(X,pr5.predict(Xp5),label= 'Poly of degree 5')
plt.plot(X,pr10.predict(Xp10),label= 'Poly of degree 10')
plt.plot(X,pr50.predict(Xp50),label= 'Poly of degree 50')
plt.plot(X,pr75.predict(Xp75),label= 'Poly of degree 75')
plt.legend();
```
