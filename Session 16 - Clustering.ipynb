```python
plt.figure(figsize=(10,8))
X,y=make_blobs(n_samples=1000, n_features=2, 
               centers=1,random_state=10)

plt.scatter(X[:,0],X[:,1]);
```

```python
plt.figure(figsize=(10,8))
X,y=make_blobs(n_samples=1000, n_features=2, 
               centers=4,random_state=10)

plt.scatter(X[:,0],X[:,1]);
```

## How to find the value of k?

  ### Elbow method

```python
from sklearn.cluster import KMeans
```

```python
SSD=[]
for k in range(1,25):
    kmeans=KMeans(n_clusters=k,random_state=10)
    kmeans.fit(X)
    SSD.append(kmeans.inertia_)
plt.plot(range(1,25),SSD);
```

From the elbow method, the optimum value of k =4

### Building the best model

```python
k_best=KMeans(n_clusters=4,random_state=10)
k_best.fit(X)
clust_pred=k_best.predict(X)
clust_pred
```

### silhouette method

```python
from sklearn.metrics import silhouette_score
SS=[]
for k in range(2,25):
    kmeans=KMeans(n_clusters=k, random_state=10)
    kmeans.fit(X)
    SS.append(silhouette_score(X,kmeans.predict(X)))
plt.plot(range(2,25), SS);
```

The peak value is 4, which is the optimum value of k.

### Clustering for iris dataset

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
# Creating a DF
X=pd.DataFrame(iris['data'],columns=['SL','SW','PL','PW'])
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

### Finding k using elbow method

```python
SSD=[]
for k in range(1,25):
    kmeans=KMeans(n_clusters=k,random_state=10)
    kmeans.fit(X_scaled)
    SSD.append(kmeans.inertia_)
plt.plot(range(1,25),SSD);
```

Choose k=3


```python
k_final=KMeans(n_clusters=3,random_state=10)
k_final.fit(X_scaled)
clus_pred=k_final.predict(X_scaled)
clus_pred
```

```python
# The cluster centriods

k_final.cluster_centers_
```
