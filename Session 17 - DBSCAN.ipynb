```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,_=make_blobs(n_samples=1000, n_features=2,
              centers=4, random_state=10)
plt.scatter(X[:,0],X[:,1]);
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=4, random_state=10)
kmeans.fit(X)
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_);
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],
           kmeans.cluster_centers_[:,1],
           c='r',marker='*',s=200);
```

## Drawbacks of KMeans

  * Spherical clustering
  
  * Value of K in advance needed
  
  * Undue influence of outliers

```python
plt.scatter(X[:,0],X[:,1]);
```

## Implementation of DBSCAN

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.5,min_samples=5)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

```python
# eps=0.4
db=DBSCAN(eps=0.4,min_samples=5)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

```python
# eps=0.3
db=DBSCAN(eps=0.3,min_samples=5)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

```python
# eps=0.2
db=DBSCAN(eps=0.2,min_samples=5)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

```python
# eps=0.05
db=DBSCAN(eps=0.05,min_samples=5)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

```python
# Fixing the best one

# eps=0.1
db=DBSCAN(eps=0.1,min_samples=4)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

## predict()

X= db.components_, the datapoints which have been clustered, taken as features

y = db.labels_ (X), taken as target

Formulate a classification problem

Choose any classification technique

```python
# Feature
X_new=db.components_
len(X_new)
```

```python
# Target
y=db.labels_[db.core_sample_indices_]
len(y)
```

## Implementing classifcation using KNN

```python
from sklearn.neighbors import KNeighborsClassifier

kn=KNeighborsClassifier(n_neighbors=20)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_new, y, test_size=0.2, random_state=10)

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Training

kn.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report

report=classification_report(y_test,kn.predict(X_test))

print(report)

```
