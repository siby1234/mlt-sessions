# Decision Tree Classifier

## Sourcing the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data set: iris dataset

```python
from sklearn.datasets import load_iris

iris=load_iris()
```

## Data preparation

```python
# Feature
X=iris['data']
X[0:10]
```

```python
# Target

y=iris['target']
y[0:10]
```

## Visualisation

```python
# Sepal length

plt.plot(X[:,0]);
```

```python
plt.plot(X[:,1]);
```

```python
plt.plot(X[:,0],label='Sepal Length')
plt.plot(X[:,1],label='Sepal Width')
plt.plot(X[:,2],label='Petal Length')
plt.plot(X[:,3],label='Petal Width')
plt.legend();
```

## Splitting the data- train and test

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,
                                              random_state=100)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train
```

## Model building

```python
from sklearn.tree import DecisionTreeClassifier

dt_1=DecisionTreeClassifier()

# Training the model

dt_1=dt_1.fit(X_train,y_train)
```

```python
dt_1
```

## Predciting using the model

```python
y_pred= dt_1.predict(X_test)
y_pred
```

```python
y_test
```

## Checking the performance

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,roc_auc_score
```

```python
cm_1=confusion_matrix(y_test,y_pred)

sns.heatmap(cm_1);
```

```python
sns.heatmap(cm_1,annot=True);
```

```python
report_1=classification_report(y_test,y_pred)
print('The classification report:\n',report_1)
```

```python
fpr,tpr,_=roc_curve(y_test,y_pred)
plt.plot(fpr,tpr);
```

## Modifying the model

```python
dt_2=DecisionTreeClassifier(criterion='entropy',max_depth=4)

dt_2=dt_2.fit(X_train,y_train)

y_pred_2=dt_2.predict(X_test)

cm_2=confusion_matrix(y_test,y_pred_2)
sns.heatmap(cm_2,annot=True);
```

```python
plot_tree(dt_2,filled=True);
```
